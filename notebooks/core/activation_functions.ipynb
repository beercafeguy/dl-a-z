# 1. tanh Activation function

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Base implementation using numpy.tanh

```python
x = np.arange(-5,5,0.25)
post_act = np.tanh(x)

plt.plot(x, post_act, color='green', linestyle='--')
plt.grid()
```

## Implement from formula

<img src="../../images/tanh_formula_1.png" alt="tanh function" style="width: 800px; height: 300px;"/>

```python
import math
def my_tanh(x):
    num = 1 - np.exp(-2 * x)
    denom = 1 + np.exp(-2 * x)
    return num / denom

x = np.arange(-5,5,0.25)
post_act = my_tanh(x)

plt.plot(x, post_act, color='red', linestyle='--')
plt.grid()
```

## Derivative of the tanh function

<img src="../../images/tanh_derivative.png" style="width: 800px; height: 200px;"/>

```python
import math
def my_tanh(x):
    num = 1 - np.exp(-2 * x)
    denom = 1 + np.exp(-2 * x)
    return num / denom

def derivative_tanh(x):
    return 1 - np.power(my_tanh(x),2)

x = np.arange(-5,5,0.25)
post_act = my_tanh(x)
tanh_derivative = derivative_tanh(x)

plt.plot(x, post_act, color='red', linestyle='--')
plt.plot(x, tanh_derivative, color='green', linestyle='-')
plt.grid()
```

Here if we notice, for higer and lower values of x, darivative just approaches 0

# 2. ReLU Activation function

<img src="../../images/relu_function.png" style="width: 400px; height: 200px;"/>

## Base implementation using numpy

```python
import math
import numpy as np
def my_relu(x):
    return np.maximum(0, x)

x = np.arange(-5,5,0.25)
post_act = my_relu(x)

plt.plot(x, post_act, color='red', linestyle='--')
plt.grid()
```

## Relu Derivative

```python
def relu_derivative(x):
    return np.where(my_relu(x) > 0, 1, 0)

def my_relu(x):
    return np.maximum(0, x)

x = np.arange(-5,5,0.25)
post_act = my_relu(x)
derivative_relu = relu_derivative(x)

plt.plot(x, post_act, color='red', linestyle='--')
plt.plot(x, derivative_relu, color='blue', linestyle='-')
plt.grid()




```
